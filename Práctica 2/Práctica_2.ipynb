1. Implemente una función en Python que permita generar una función escalón unitario, definida en un intervalo 𝑛1≤𝑛0≤𝑛2.

```python
def escalon_unitario (n0, n1, n2):
    m = np.arange(n1, n2 + 1)
    c = (m >= n0).astype(int)
    return c,m
```

```python
m = escalon_unitario(0,-2,5)[0]
c = escalon_unitario(0,-2,5)[1]

print(f'n = {c}\n x = {m}')
```

```python
plt.stem(c,m)
plt.xlabel('n', size= "15", fontname='Times New Roman')
plt.ylabel('x(n)', size= "15", fontname='Times New Roman')
plt.title('Representación Escalón Unitario (secuencia discreta)',size= "20", fontname='Times New Roman',color='red')
plt.grid(True)
```

2. Implemente una función en Python que permita generar una función rampa, definida en un intervalo 𝑛1≤𝑛0≤𝑛2 con una pendiente m. 

```python
def rampa (n0, n1, n2):
    w = np.arange(n1, n2 + 1)
    y = np.where(w>=0, w-n0, 0).astype(int)
     
    return y,w
```

```python
w = rampa(0,-2,5)[0]
y = rampa(0,-2,5)[1]

print(f'n = {y}\n x = {w}')
```

```python
plt.stem(y,w)
plt.xlabel('n', size= "15", fontname='Times New Roman')
plt.ylabel('x(n)', size= "15", fontname='Times New Roman')
plt.title('Representación Rampa (secuencia discreta)', size= "20", fontname='Times New Roman',color='red')
plt.grid(True)
```

3. Implemente una función en Python que permita generar una función exponencial compleja, definida en un intervalo 𝑛1≤𝑛0≤𝑛2. 

```python
def exponencial_compleja (sigma, w0, n1, n2):
    l = np.arange(n1, n2 + 1)
    b = np.exp(sigma + 1j*w0)*l

    return l,b
```

```python
sigma = -8
w0 = 3
n1 = 0
n2  = 13

l = exponencial_compleja (sigma,w0,n1,n2)[0]
b = exponencial_compleja (sigma,w0,n1,n2)[1]

print(f'n = {l}\n x = {b}')
```

```python
plt.stem(l,b)
plt.xlabel('n', size= "15", fontname='Times New Roman')
plt.ylabel('x(n)', size= "15", fontname='Times New Roman')
plt.title('Representación Exponencial Compleja (secuencia discreta)', size= "20", fontname='Times New Roman',color='red')
plt.grid(True)
```

4. Implemente una función en Python que permita generar una función sinusoidal, definida en un intervalo 𝑛1≤𝑛0≤𝑛2. 

```python
def sinusoide (w0,theta, n1, n2):
    z = np.arange(n1, n2 + 1)
    p = np.sin(w0*z + theta)

    return z, p
```

```python
theta = 8
w0 = 2*np.pi/12
n1 = 0
n2 = 13

z = sinusoide (w0,theta,n1,n2)[0]
p = sinusoide (w0,theta, n1,n2)[1]

print(f'n = {z}\n x = {p}')
```

```python
plt.stem(z,p)
plt.xlabel('n', size= "15", fontname='Times New Roman')
plt.ylabel('x(n)', size= "15", fontname='Times New Roman')
plt.title('Representación Sinusoide (secuencia discreta)', size= "20", fontname='Times New Roman',color='red')
plt.grid(True)
```

```python
#Para a:

import numpy as np
import matplotlib.pyplot as plt

# Impulso discreto δ(n)
def delta(n):
    return np.where(n == 0, 1, 0)

# Escalón unitario u(n)
def u(n):
    return np.where(n >= 0, 1, 0)

# Rampa r(n)
def r(n):
    return n * u(n)
    
n = np.arange(-4,3)

x1 = 3*delta(n+1) + 5*delta(n+3) + 3*delta(n+2) + 3*delta(n) + delta(n)

plt.stem(n, x1)
plt.title("Secuencia x1(n)",  color = "orange" , size = "20", fontname='Times New Roman')
plt.xlabel("n", color = "green" , size = "15", fontname='Times New Roman')
plt.ylabel("Amplitud", color = "green" , size = "15", fontname='Times New Roman')
plt.grid()
plt.show()
```

```python
#Para b:
n = np.arange(-10,11)

x4 = 2*r(n+3) - r(n-2) - 5*u(n-3)

plt.stem(n, x4)
plt.title("Secuencia x4(n)",  color = "orange" , size = "20", fontname='Times New Roman')
plt.xlabel("n",  color = "green" , size = "15", fontname='Times New Roman')
plt.ylabel("Amplitud", color = "green" , size = "15", fontname='Times New Roman')
plt.grid()
plt.show()
```

6. Genere la siguiente secuencia 

x[𝑛] = r[n] ,0 ≤𝑛≤5

𝑟[𝑛 −5], 6 ≤𝑛≤11

𝑟[𝑛 −10]   12 ≤𝑛 ≤17

a. Derive la secuencia  

b. Realice un subplot con las dos secuencias, la original, y la derivada. Describa lo que observa
y el porqué no es exactamente la derivada del caso continuo.


```python
#Para generar la secuencia por tramos:

n = np.arange(0,18)

x = np.zeros(len(n))

for i in range(len(n)):
    
    if 0 <= n[i] <= 5:
        x[i] = r(n[i])
        
    elif 6 <= n[i] <= 11:
        x[i] = r(n[i]-5)
        
    elif 12 <= n[i] <= 17:
        x[i] = r(n[i]-10)

dx = np.diff(x, prepend=0) #prepend mantiene el tamaño

plt.figure(figsize=(10,8))

plt.subplot(2,1,1)
plt.stem(n,x)
plt.title("Secuencia original", color = "orange" , size = "20", fontname='Times New Roman')
plt.xlabel("n",color ="green", size= "15", fontname='Times New Roman')
plt.ylabel("x[n]",color ="green", size= "15", fontname='Times New Roman')
plt.grid()

plt.subplot(2,1,2)
plt.stem(n,dx)
plt.title("Derivada discreta", color = "orange" , size = "20", fontname='Times New Roman')
plt.xlabel("n",color ="green", size= "15", fontname='Times New Roman')
plt.ylabel("dx[n]", color ="green", size= "15", fontname='Times New Roman')
plt.grid()

plt.tight_layout()
plt.show()
```

En la gráfica superior se observa la secuencia original 𝑥[𝑛], la cual corresponde a una rampa discreta creciente que aumenta aproximadamente una unidad en cada muestra. Esta rampa se repite periódicamente cada 6 muestras aproximadamente, lo que indica que después de alcanzar un valor máximo la señal se reinicia y vuelve a comenzar desde un valor bajo. Este comportamiento genera una señal tipo rampa periódica o diente de sierra discreto.

En la gráfica inferior se muestra la derivada discreta 𝑑𝑥[𝑛] obtenida mediante diferencias entre muestras consecutivas. En los intervalos donde la rampa crece de forma uniforme, la derivada presenta valores constantes cercanos a 1, lo cual es coherente con la pendiente de la rampa, ya que la señal aumenta aproximadamente una unidad por muestra.

Sin embargo, en los puntos donde la señal se reinicia abruptamente, se observan picos negativos pronunciados en la derivada. Esto ocurre porque la señal pasa repentinamente de un valor alto a uno mucho menor, generando una diferencia negativa grande entre muestras consecutivas. Estos picos representan el cambio brusco o discontinuidad en la señal original.

Es importante notar que la derivada discreta no es exactamente igual a la derivada en tiempo continuo, debido a que en sistemas discretos:

- No existe una derivada exacta como en el caso continuo.

- Se utiliza una aproximación mediante diferencias finitas entre muestras.

- La presencia de saltos o discontinuidades en la señal genera valores extremos en la derivada.

Por lo tanto, la derivada discreta refleja correctamente la tendencia de crecimiento de la señal y los cambios abruptos, aunque introduce pequeñas diferencias respecto al comportamiento ideal de una derivada continua.

```python
# partes de la secuencia
a = np.arange(0,6)      # 0 a 5
b = np.arange(4,-1,-1)  # 4 a 0
c = np.arange(1,6)      # 1 a 5
d = np.ones(5)*5        # cinco 5
e = np.ones(4)*10       # cuatro 10

# concatenar
x = np.concatenate((a,b,c,d,e))

print(f'Secuencia x(n) = {x}')

n = np.arange(0,len(x))

plt.stem(n,x)
plt.title("Secuencia x(n)", color = "orange" , size = "20", fontname='Times New Roman')
plt.xlabel("n", color = "blue" , size = "15", fontname='Times New Roman')
plt.ylabel("Amplitud", color = "blue" , size = "15", fontname='Times New Roman')
plt.grid()
plt.show()
```
